```python
import pandas as pd 
import numpy as np 

import warnings 
warnings.filterwarnings('ignore')
```

```python
tatamotor_df= pd.read_csv('tatamotor.csv')
tatamotor_df=tatamotor_df[['Date', 'Close Price']]
tatamotor_df.dtypes
```

```python
mahindra_df= pd.read_csv('mahindra.csv')
mahindra_df=mahindra_df[['Date', 'Close Price']]
mahindra_df.dtypes
```

```python
tatamotor_df['Date']= pd.DatetimeIndex(tatamotor_df['Date'])
```

```python
tatamotor_df
```

```python
mahindra_df['Date']= pd.DatetimeIndex(mahindra_df['Date'])
mahindra_df.dtypes
```

```python
tatamotor_df=tatamotor_df.set_index('Date')
```

```python
tatamotor_df
```

```python
mahindra_df=mahindra_df.set_index('Date')
```

```python
mahindra_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(15,5))
plt.plot(tatamotor_df)
```

```python
plt.figure(figsize=(15,5))
plt.plot(mahindra_df)
```

```python
tatamotor_df['gain']=tatamotor_df['Close Price'].pct_change(periods=1)
```

```python
tatamotor_df
```

```python
mahindra_df['gain']=mahindra_df['Close Price'].pct_change(periods=1)
```

```python
mahindra_df
```

```python
plt.figure(figsize=(15,5))
plt.plot(tatamotor_df.gain , label="Tata_Motor")
plt.plot(mahindra_df.gain, label= "Mahindra")
plt.legend()
```

```python
plt.figure(figsize=(15,5))
plt.plot(mahindra_df.gain)
```

```python
sns.distplot(tatamotor_df.gain, label="Tata Motor")
sns.distplot(mahindra_df.gain, label= "Mahindra")
plt.legend()
```

```python
print("Mean:", round(tatamotor_df.gain.mean(),4))
print("Std:",round(tatamotor_df.gain.std(),4))
```

```python
print("Mean:", round(mahindra_df.gain.mean(),4))
print("Std:",round(mahindra_df.gain.std(),4))
```

```python
tatamotor_df.dropna()
mahindra_df.dropna()
```
